```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(0)

# Time vector
t = np.linspace(0, 10, 500)

# First random signal
signal_1 = np.random.randn(len(t))

# Second random signal
signal_2 = np.random.randn(len(t))

# Constant signal
constant_signal = np.full_like(t, 0.5)

# Sum of the signals (first + second + constant)
sum_signal = signal_1 + signal_2 + constant_signal

# Create a figure and set up subplots
fig, axs = plt.subplots(4, 1, figsize=(10, 20))  # 4 subplots in a single column

# First plot (Random Signal 1)
axs[0].plot(t, signal_1, 'k-', linewidth=2)
axs[0].set_title("Random Signal 1")
axs[0].set_xlabel('t')
axs[0].set_ylabel(r'$\xi_1(t)$')
axs[0].tick_params(left=False, right=False, labelleft=False, labelbottom=False)

# Second plot (Random Signal 2 with tab)
axs[1].plot(t, signal_2, 'k-', linewidth=2)
axs[1].set_title("Random Signal 2")
axs[1].set_xlabel('t')
axs[1].set_ylabel(r'$\xi_1(t)$')
axs[1].tick_params(left=False, right=False, labelleft=False, labelbottom=False)

# Third plot (Constant Signal)
axs[2].plot(t, constant_signal, 'k-', linewidth=2)
axs[2].set_title("Constant Signal")
axs[2].set_xlabel('t')
axs[2].set_ylabel(r'$\xi_1(t)$')
axs[2].tick_params(left=False, right=False, labelleft=False, labelbottom=False)

# Fourth plot (Sum of all previous signals)
axs[3].plot(t, sum_signal, 'k-', linewidth=2)
axs[3].set_title("Sum of Signals")
axs[3].set_xlabel('t')
axs[3].set_ylabel(r'$\xi_1(t)$')

# Adjust layout for better spacing between plots
plt.tight_layout(h_pad=5.0)

# Show the plots
plt.show()



```
